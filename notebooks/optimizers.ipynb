# Optimizers

Optimizers in deep learning are algorithms that adjust the model’s parameters during training to minimize a loss function. They enable neural networks to learn from data by iteratively updating weights and biases. 

Common optimizers:
1. Stochastic Gradient Descent (SGD)
2. SGD with momentum
3. Adaptive Moment Estimation (Adam)
4. Root Mean Square Propagation (RMSProp)

## 1. Stochastic Gradient Descent (SGD)

SGD is the most basic optimizer. It updates the weights and biases in the direction of the negative gradient of the loss function with respect to the weights and biases. The learning rate determines the size of the steps taken in the direction of the negative gradient.

Equation for SGD:

$$
\theta_{t+1} = \theta_t - \eta \nabla L(\theta_t)
$$

where $\theta$ is the model’s parameters, $\eta$ is the learning rate, and $\nabla L(\theta_t)$ is the gradient of the loss function with respect to the model’s parameters.

```python
# example
import numpy as np
import matplotlib.pyplot as plt

# Generate some random data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Plot the data
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Generated Data')
plt.show()

```

```python
# Add a bias term to the feature matrix
X_b = np.c_[np.ones((100, 1)), X]

# Initialize random values for the model parameters
theta = np.random.randn(2, 1)

# Set hyperparameters
learning_rate = 0.01
n_iterations = 1000

# SGD algorithm
for iteration in range(n_iterations):
    # Randomly shuffle the data
    shuffled_indices = np.random.permutation(100)
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    
    # Iterate through each data point
    for i in range(100):
        xi = X_b_shuffled[i:i+1]
        yi = y_shuffled[i:i+1]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        theta = theta - learning_rate * gradients

# Print the final parameters
print("Final theta:", theta)
```

```python
# Plot the regression line
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Generated Data')
plt.plot(X, X_b.dot(theta), color='red')
plt.show()
```
